```python

df['room_layout'] = df['room_quantity'].astype(str).str.replace(r'[^А-Яа-я]','', regex=True)
df['room_layout'] = df['room_layout'].fillna('не указано')

df['room_layout']=df['room_layout'].str.strip()
df['room_layout'] =df['room_layout'].fillna('не указано ')
df['room_layout']
```

```python
df['room_count'] = df['room_quantity'].astype(str).str.replace(r'[^0-9.]', '', regex=True)
df['room_count'] = pd.to_numeric(df['room_count'], errors='coerce')
mean_room = df['room_count'].mean()
df['room_count'] = df['room_count'].fillna(mean_room)
df['room_count'] =df['room_count'].astype(int)
df['room_count']
```

```python
df['square_meter'] = df['square_meter'].str.split('/').str[0]
df['square_meter'] = pd.to_numeric(df['square_meter'], errors='coerce').fillna(0).astype(float)
df['square_meter']
```

```python

# df['number_of_floors'] = df['house_number'].astype(str).replace(r'[А-Яа-я]', '', regex=True)
# df['number_of_floors'] = pd.to_numeric(df['number_of_floors'], errors='coerce')
# df['number_of_floors']

df['number_of_floors'] = df['house_number'].str.split(',').str[0]
# df['number_of_floors'].unique()
df['type_of_house'] = df['house_number'].str.split(',').str[1]
df['type_of_house'] = df['type_of_house'].fillna('не указано')
df['type_of_house']
df['number_of_floors'] = df['number_of_floors'].fillna(0)
df['number_of_floors']
```

```python
 
def format_phone(number):
    if not number or pd.isna(number):
        return number
    num = "".join(filter(str.isdigit, str(number)))
    
    if len(num) == 12 or (len(num) == 11 and str(number).startswith('+')):
        return f'+7({num[-10:-7]}){num[-7:-4]}-{num[-4:-2]}-{num[-2:]}'
    return number
df['phone_number'] =  df['phone_number'].str.split(',').apply(
    lambda x: ','.join([format_phone(i.strip()) for i in x]) if isinstance(x, list) else x
)
df['phone_number']


```

```python
df['balcony'] = df['balcony'].fillna('не указано')
df['balcony'].value_counts()

```

```python
df[['windows','bathroom','children_pets_allowed', 'residential complex', 'serial_house']]=df[['windows','bathroom','children_pets_allowed', 'residential complex', 'serial_house']].fillna('не указано')
df.head(4)      
```

```python
df['height'].mean(),df['height'].median()
median_high = df['height'].median()
df['height'] = df['height'].fillna(median_high)
df['height']
```

```python
df['m2_room'] = df['m2_room'].fillna('не указано')
df['garbage_chute'] = df['garbage_chute'].fillna('не указано')	
df.head(4)
```

```python
df.loc[df['height'] > 6,'height'] = np.nan
median_height = df['height'].median()
df['height'] = df['height'].fillna(median_height)
df['height']
```

```python
fig = px.scatter(df, x='square_meter', y='price_per_month',
                trendline='ols',
                title='Связь цены аренды и общей площади',
                labels={'square_meter':'Площадь кв.М', 'price_per_month':'Цена аренда квартиры за месяц'},
                hover_data=['room_count','subway'])
fig.show()
```

```python
df['number_of_floors'] = df['number_of_floors'].astype(str).str.split('/').str[0]
df.loc[df['number_of_floors'] == '-1', 'number_of_floors'] = np.nan
df['number_of_floors'] = pd.to_numeric(df['number_of_floors'],errors='coerce')

num_floor= df['number_of_floors'].median()
df['number_of_floors'] = df['number_of_floors'].fillna(num_floor)
df['number_of_floors'] = df['number_of_floors'].astype('Int64')



df['number_of_floors'].info()
filtered_floors = df[df['number_of_floors'] <= 25]
table = filtered_floors.pivot_table(
    index='number_of_floors',
    columns='lift',
    values='price_per_month',
    aggfunc='count'
).fillna(0)

plt.figure(figsize=(10,12))
sns.heatmap(table, annot=True, fmt='.0f', cmap='Blues', cbar_kws={'label': 'Количество обьявлений'})

plt.title('Техю Состояние: Взаимосвяз Этажности и наличия лифта')
plt.xlabel('Тип лифта', fontsize=12)
plt.ylabel('Этажность дома', fontsize=12)
plt.show()
```
